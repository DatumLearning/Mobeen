```
import numpy as np
```

```
means = [28 , 21 , 15 , 7 , 2]
sd = [2.5 , 2 , 1.5 , 1 , 0.5]
week_study = []
for i in  range(5):
  num = np.random.normal(means[i] , sd[i] , 200)
  num[num<0] = 0
  num = np.floor(num)
  week_study.extend(num)
```

```

probs = [0.9 , 0.7 , 0.5 , 0.3 , 0.1]
poss = [1 , 0]
school = []
for i in range(5):
    sch = np.random.choice(poss , 200 , p = [probs[i] , 1 - probs[i]])
    school.extend(sch)
```

```
import pandas as pd
train = pd.DataFrame(data)
```
